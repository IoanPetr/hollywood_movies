```python
import numpy as np

# Normalized input (height)
x = np.array([-0.5427, -0.4093, -0.0760, 0.1240, 0.1907, 0.2573, 0.4573])

# Normalized target (weight)
y = np.array([-0.5106, -0.3015, -0.2460, 0.0318, 0.2124, 0.3235, 0.4901])

m = len(x)

# Learning rate
alpha = 0.1

# Initialize parameters
theta0 = 0.0
theta1 = 0.0

def compute_cost(theta0, theta1, x, y):
    h = theta0 + theta1 * x
    return (1/(2*m)) * np.sum((h - y)**2)

print("Iter |    theta0    |    theta1    |     Cost")
print("-------------------------------------------------")

for i in range(1, 11):  # 1 to 50
    # Prediction
    h = theta0 + theta1 * x

    # Gradients
    grad0 = (1/m) * np.sum(h - y)
    grad1 = (1/m) * np.sum((h - y) * x)

    # Update
    theta0 -= alpha * grad0
    theta1 -= alpha * grad1

    # Print iteration values
    J = compute_cost(theta0, theta1, x, y)
    print(f"{i:4d} | {theta0:12.6f} | {theta1:12.6f} | {J:12.6f}")
```

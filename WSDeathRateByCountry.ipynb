```python
import requests
import pandas as pd
from io import StringIO
from google.colab import files  # for downloading in Colab

# 1️⃣ Target URL
url = "https://worldpopulationreview.com/country-rankings/death-rate-by-country"

# 2️⃣ Headers with correct ASCII hyphen
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

# 3️⃣ Fetch the page
resp = requests.get(url, headers=headers)
html = resp.text

# 4️⃣ Read tables with pandas
tables = pd.read_html(StringIO(html))
print(f"Found {len(tables)} table(s)")

# 5️⃣ Clean first table
df = tables[0]
# Remove weird characters from column names
df.columns = df.columns.str.encode('ascii', 'ignore').str.decode('utf-8')
print(df.head())

# 6️⃣ Save to CSV
csv_filename = "death_rate_by_country.csv"
df.to_csv(csv_filename, index=False, encoding="utf-8")
print(f"Saved table to {csv_filename}")

# 7️⃣ Trigger download in Colab
files.download(csv_filename)

```

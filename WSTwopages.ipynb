```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# --- Σελίδα 1: Quotes to Scrape ---
quotes_url = 'https://quotes.toscrape.com/'
quotes_data = []

while quotes_url:
    response = requests.get(quotes_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    for quote in soup.select('div.quote'):
        quotes_data.append({
            'type': 'quote',
            'text': quote.select_one('span.text').get_text(),
            'author': quote.select_one('small.author').get_text(),
            'price': None  # Κενό για συμβατότητα με books
        })

    next_page = soup.select_one('li.next a')
    quotes_url = 'https://quotes.toscrape.com' + next_page['href'] if next_page else None

# --- Σελίδα 2: Books to Scrape ---
books_url = 'https://books.toscrape.com/catalogue/page-1.html'
books_data = []

while books_url:
    response = requests.get(books_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    for book in soup.select('article.product_pod'):
        books_data.append({
            'type': 'book',
            'text': None,  # Κενό για συμβατότητα με quotes
            'author': None,  # Κενό για συμβατότητα με quotes
            'title': book.select_one('h3 a')['title'],
            'price': book.select_one('p.price_color').get_text()
        })

    next_page = soup.select_one('li.next a')
    books_url = 'https://books.toscrape.com/catalogue/' + next_page['href'] if next_page else None

# --- Συνένωση δεδομένων και αποθήκευση σε CSV ---
# Δημιουργούμε DataFrame
df_quotes = pd.DataFrame(quotes_data)
df_books = pd.DataFrame(books_data)

# Συγχώνευση σε ένα DataFrame (προαιρετικά μπορούμε να κρατήσουμε ξεχωριστά)
df = pd.concat([df_quotes, df_books], ignore_index=True)

# Αποθήκευση σε CSV
df.to_csv('scraped_data.csv', index=False, encoding='utf-8-sig')

print("Τα δεδομένα σώθηκαν σε 'scraped_data.csv'")
df.head(50)  # Δείχνει τα πρώτα 10 στοιχεία

from google.colab import files

# Κατεβάζουμε το CSV
files.download('scraped_data.csv')
```
